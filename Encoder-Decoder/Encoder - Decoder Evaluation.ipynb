```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import accuracy_score
```

```python
languages = list(sorted(os.listdir("openmt_data/")))
```

```python
accuracies = []

for lang in languages:
    pred_file = f"predictions/{lang}.pred"
    true_file = f"openmt_data/{lang}/test.trg"
    
    if os.path.exists(pred_file) and os.path.exists(true_file):
        with open(pred_file) as pf:
            y_pred = [pred.strip() for pred in pf]
        
        with open(true_file) as tf:
            y_true = [form.strip() for form in tf]
        
        accuracy = accuracy_score(y_pred=y_pred, y_true=y_true)
    
    else:
        accuracy = 0.0
    
    accuracies.append(accuracy)
```

```python
results = pd.DataFrame(np.array([languages, accuracies]).T, columns=["Language", "Accuracy"])
```

```python
results.sort_values('Accuracy', ascending=False, inplace=True)

accuracies = results['Accuracy'].astype(np.float64)
languages = results['Language']

plt.figure(figsize=(16, 4))
plt.bar(languages, accuracies)
plt.show()
```
